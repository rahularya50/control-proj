```python
from linalg_helper import make_random_direction

import numpy as np

dim = 10

def make_Ds(dim, num_ds):
    Ds = []
    for i in range(num_ds):
        a = make_random_direction(dim)
        Ds.append(np.outer(a, a))
    return Ds

def trials(N, dim):
    K = 10.0*np.random.randn(dim, dim)
    Ds = make_Ds(dim, N)
    Dsum = np.zeros([dim, dim])
    for D in Ds:
        Dsum += D @ K
    avg = Dsum/N
    expected = K * 1/dim
    print(avg)
    print(expected)
    
trials(100000, dim)
```

```python
def make_sqsq_avg(dim, num):
    sqsq = []
    for i in range(num):
        a = make_random_direction(dim)
        sqsq.append(a[0] ** 2 * a[1] ** 2)
    return np.mean(sqsq)

def make_cubt_avg(dim, num):
    cubt = []
    for i in range(num):
        a = make_random_direction(dim)
        cubt.append(a[0] ** 3 * a[1])
    return np.mean(cubt)
```

```python
def make_sqtt_avg(dim, num):
    sqtt = []
    for i in range(num):
        a = make_random_direction(dim)
        sqtt.append(a[0] ** 2 * a[1] * a[2])
    return np.mean(sqtt)
```

```python
def make_fourth_avg(dim, num):
    fourths = []
    for i in range(num):
        a = make_random_direction(dim)
        fourths.append(a[0] ** 4)
    return np.mean(fourths)
```
